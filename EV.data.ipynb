```python
# visualization of Electric Range From  1997-2024
sns.displot(x='Electric Range',data=df)
```

```python
#Visualization of Base MSRP by make from 1998-2024

plt.figure(figsize=(10,5))
sns.barplot(y='Base MSRP',x='Make',data=df)
plt.title("Base MSRP vs Make")
plt.xticks(rotation=90)
```

```python
# Top most  car models of  electric vehicle from 1997-2024.

top_most=df['Model'].value_counts().sort_values(ascending=False,).reset_index().head(10)

sns.barplot(y='index',x='Model',data=top_most)

plt.xlabel('Model')

plt.ylabel('No_of_models')

plt.title('Top most  car models of  electric vehicle from 1997-2024. ')
```

```python
# showing visualization of company located in legislative district of sate washingston

plt.figure(figsize=(10,5))
sns.boxplot(y="Legislative District",x="Make",data=df)
plt.xticks(rotation=90)
```

```python
# top most electric vehicles companies from 1997-2024.

top_most=df['Make'].value_counts().sort_values(ascending=False,).reset_index().head()

sns.barplot(y='Make',x='index',data=top_most)
plt.title("top most manufacturing electric vehicles companies from 1997-2024")
plt.xlabel('Top Companies')
plt.ylabel('No_of_makes')
```

```python
# visualization of models which is made by tesla from 1997 to 2024

Tesla=df[df['Make']=='TESLA']
sns.stripplot(x='Model',y='Make',data=Tesla)

plt.title("Models of Telsa")
plt.ylabel('company')
plt.xlabel('Models')
```

```python
# visualization of Electric Vehicle type from 1997-2024

sns.countplot(x='Electric Vehicle Type',data=df)

plt.title('Types of Electric Vehicles from 1997-2024')

plt.ylabel('Numbber of Electric vehicle type')
```

```python
# visualization of electric range used by companies from 1997-2024

sns.stripplot(x='Make',y='Electric Range',data=df)
plt.xticks(rotation=90)
```

```python
# visualization of model year which made by companies(make) from 1997-2024

plt.figure(figsize=(10,5))
sns.violinplot(y='Model Year',x='Make',data=df)
plt.xticks(rotation=90)
```

```python
# showing the relationship between the longitude and latiude

sns.scatterplot(y='Longitude',x='latitude',data=df)
```
